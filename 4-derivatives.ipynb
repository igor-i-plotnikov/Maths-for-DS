### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
from math import sin, cos
from scipy.misc import derivative
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
```

```python
x = 1
```

```python
derivative(f, x, dx=1e-6, n=1)
```

```python
derivative(f, x, dx=1e-6, n=2)
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return sin(2*x + 1)**5
```

```python
def der1(x):
    return derivative(f, x, dx=1e-6, n=1)
```

```python
def der2(x):
    return derivative(f, x, dx=1e-6, n=2)
```

```python
xs = np.linspace(-5, 5, num=500)
plt.plot(xs, [f(x) for x in xs], color='green', label='Function')
plt.plot(xs, [der1(x) for x in xs], color='blue', label='Derivative1')
plt.plot(xs, [der2(x) for x in xs], color='purple', label='Derivative2')
```
