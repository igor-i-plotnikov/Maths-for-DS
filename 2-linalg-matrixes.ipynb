### Уровень 1:
Дан объект в 2d пространстве:

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи матричных операций выполните:

- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x
- Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)

```python
reduced = A/2
x, y = reduced[:,0], reduced[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
from math import sin, cos
```

```python
def rotate_matrix(matrix, angle_in_deg):
    theta = np.deg2rad(angle_in_deg)
    rot = np.array([[cos(theta), sin(theta), 0],[-sin(theta), cos(theta), 0],[0, 0, 1]])
    return matrix @ rot
```

```python
rotated = rotate_matrix(A, 130)
x, y = rotated[:,0], rotated[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
reflected = A @ np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
x, y = reflected[:,0], reflected[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
translated = np.array([(v[0] + 200, v[1] + 300, v[2]) for v in A])
x, y = translated[:,0], translated[:,1]
plt.plot(x, y)
plt.ylim([-200, 500])
plt.xlim([-200, 400])
```

```python
scaled = np.array([(v[0] * 0.5, v[1] * 1.2, v[2]) for v in A])
x, y = scaled[:,0], scaled[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
